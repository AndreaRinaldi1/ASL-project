```python
import numpy as np
```

```python
path = "/home/andrea/Desktop/Baseline6Prime/Write_S3_MW2_W8/"
files = [[],[],[]]
for vm in range(1, 4):
    for instance in range(1, 3):
        for rep in range(1, 4):
            files[rep-1].append(path+"VM"+str(vm)+"_"+str(instance)+"/rep"+str(rep)+"/output.log")
            
```

```python
latency = np.zeros(4)
throughput = np.zeros(4)
for rep in range(1, 4):
     
    for j in range(0, len(files[rep-1])):
        fileI = open(files[rep-1][j], "r")
        linesI = []
        for line in fileI:
            linesI.append(line)

        xI = np.asarray(linesI)
        

        countI = 0
        i = 2
        while xI[i] != "ALL STATS\n":
            countI += 1
            i += 1

        i += 4  #4 per set, 5 per get, 7 per 50%50
        z = xI[i][6:]
        m = 0
        numbers = 0
        singleLatency = ""
        singleThroughput = ""
        while numbers < 3:
            while z[m] == " ":
                m += 1
            while z[m] != " ":
                if numbers == 0:
                    singleThroughput += z[m]
                m += 1 
            numbers += 1
        while z[m] == " ":
                m += 1
        while z[m] != " ":
            singleLatency += z[m]
            m += 1

        latency[rep-1] += float(singleLatency)
        throughput[rep-1] += float(singleThroughput)
        
    latency[rep-1] = latency[rep-1] / 6

throughput[3] = np.mean(throughput[0:3])
print((throughput[0:3] - throughput[3])**2)
latency[3] = np.mean(latency[0:3])
```

```python
fileOutput = open(path+"results.txt", "w")
```

```python
fileOutput.write("Servers: 3\tMiddleware: 2\tWorker Threads: 8\n")
fileOutput.write("Throuhgput: ")
fileOutput.write(str(throughput[0])+ " " + str(throughput[1]) + " " + str(throughput[2]))
fileOutput.write("\nAverage Throughput: ")
fileOutput.write(str(throughput[3]))
fileOutput.write('\nResponse Time: ')
fileOutput.write(str(latency[0])+ " " + str(latency[1]) + " " + str(latency[2]))
fileOutput.write("\nAverage Response Time: ")
fileOutput.write(str(latency[3]))

fileOutput.close()
```
